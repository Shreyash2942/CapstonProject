```redshift
--Which disease has the maximum number of claims
SELECT disease_name, COUNT(*) AS claim_count
FROM capstonproject.claims
GROUP BY disease_name
ORDER BY claim_count DESC;
```

```redshift
--Which group has the maximum subgroups

SELECT Grp_Id, COUNT(DISTINCT SubGrp_ID) AS subgroup_count
FROM capstonproject.grpsubgrp
GROUP BY Grp_Id
ORDER BY subgroup_count DESC;
```

```redshift
--Find out the hospital which served the most number of patients
SELECT h.Hospital_name, COUNT(*) AS patient_count
FROM capstonproject.patient_records p
JOIN capstonproject.hospital h ON p.hospital_id = h.Hospital_id
GROUP BY h.Hospital_name
ORDER BY patient_count DESC;
```

```redshift
--Assuming 'N' means rejected in the Claim_Or_Rejected column.
SELECT COUNT(*) AS rejected_claims
FROM capstonproject.claims
WHERE Claim_Or_Rejected = 'N';
```

```redshift
--Which group is most profitable
SELECT Grp_Id, Grp_Name, premium_written
FROM capstonproject.group
ORDER BY premium_written DESC;
```

```redshift
-- Find out which subgroups are subscribed the most number of times
SELECT 
    s.Subgrp_id,
    sg.SubGrp_Name,
    COUNT(*) AS subscription_count
FROM capstonproject.subscriber s
JOIN capstonproject.subgroup sg ON s.Subgrp_id = sg.SubGrp_id
GROUP BY s.Subgrp_id, sg.SubGrp_Name
ORDER BY subscription_count DESC;
```

```redshift
--Which group is most profitable
SELECT 
    g.Grp_Id,
    grp.Grp_Name,
    COUNT(DISTINCT g.SubGrp_ID) AS subgroup_count
FROM capstonproject.grpsubgrp g
JOIN capstonproject.group grp ON g.Grp_Id = grp.Grp_Id
GROUP BY g.Grp_Id, grp.Grp_Name
ORDER BY subgroup_count DESC;
```
