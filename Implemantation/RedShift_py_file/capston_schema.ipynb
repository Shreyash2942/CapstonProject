```redshift
CREATE SCHEMA capstonproject
```

```redshift
CREATE TABLE capstonproject.claims (
    Claim_Or_Rejected VARCHAR(5),
    SUB_ID VARCHAR(50),
    claim_amount NUMERIC(12, 2),
    claim_date VARCHAR(50),
    claim_id INT,
    claim_type VARCHAR(50),
    disease_name VARCHAR(100),
    patient_id INT
    
);
```

```redshift
CREATE TABLE capstonproject.grpsubgrp (
    SubGrp_ID VARCHAR(50),
    Grp_Id VARCHAR(50)
);
```

```redshift
CREATE TABLE capstonproject.group (
    Country VARCHAR(100),
    premium_written INTEGER,
    zipcode INTEGER,
    Grp_Id VARCHAR(50),
    Grp_Name VARCHAR(255),
    Grp_Type VARCHAR(50),
    city VARCHAR(100),
    year INTEGER
);
```

```redshift
CREATE TABLE capstonproject.disease (
    SubGrpID VARCHAR(50),
    Disease_ID INTEGER,
    Disease_name VARCHAR(255)
);
```

```redshift
CREATE TABLE capstonproject.hospital (
    Hospital_id VARCHAR(50),
    Hospital_name VARCHAR(255),
    city VARCHAR(100),
    state VARCHAR(100),
    country VARCHAR(100)
);
```

```redshift
CREATE TABLE capstonproject.patient_records (
    Patient_id INTEGER,
    Patient_name VARCHAR(100),
    patient_gender VARCHAR(10),
    patient_birth_date DATE,
    patient_phone VARCHAR(20),
    disease_name VARCHAR(100),
    city VARCHAR(100),
    hospital_id VARCHAR(50)
);
```

```redshift
CREATE TABLE capstonproject.subgroup (
    SubGrp_id VARCHAR(50),
    SubGrp_Name VARCHAR(255),
    Monthly_Premium INTEGER
);
```

```redshift
CREATE TABLE capstonproject.subscriber (
    sub_id VARCHAR(50),
    first_name VARCHAR(100),
    last_name VARCHAR(100),
    Street VARCHAR(255),
    Birth_date DATE,
    Gender VARCHAR(10),
    Phone VARCHAR(20),
    Country VARCHAR(100),
    City VARCHAR(100),
    Zip_Code INTEGER,
    Subgrp_id VARCHAR(50),
    Elig_ind VARCHAR(5),
    eff_date DATE,
    term_date DATE
);
```
